```python
import sqlite3
```

```python
conn = sqlite3.connect('muvi_database')
```

```python
c = conn.cursor()
```

```python
c.execute('CREATE TABLE IF NOT EXISTS casts_replace (id INTEGER PRIMARY KEY AUTOINCREMENT, movie_id INTEGER, cast_id INTEGER, character TEXT, credit_id INTEGER, actor_id INTEGER, order_number INTEGER)')
c.execute('CREATE TABLE IF NOT EXISTS actors_replace (id INTEGER PRIMARY KEY AUTOINCREMENT, gender INTEGER, name TEXT, profile_path TEXT)')
c.execute('CREATE TABLE IF NOT EXISTS collections_replace (id INTEGER PRIMARY KEY AUTOINCREMENT, collection_id INTEGER, name TEXT, poster_path TEXT, backdrop_path TEXT)')
c.execute('CREATE TABLE IF NOT EXISTS movies_replace (id INTEGER PRIMARY KEY AUTOINCREMENT, adult BOOL, belongs_to_collection INTEGER, budget INTEGER, homepage TEXT, movie_id INTEGER, imdb_id TEXT, original_language TEXT, original_title TEXT, overview TEXT, popularity REAL, poster_path TEXT, release_date TEXT, revenue INTEGER, runtime REAL, status TEXT, tagline TEXT, title TEXT, vote_average REAL, vote_count INTEGER)')
c.execute('CREATE TABLE IF NOT EXISTS genres_replace (id INTEGER PRIMARY KEY AUTOINCREMENT, name TEXT)')
c.execute('CREATE TABLE IF NOT EXISTS movieGenres_replace (id INTEGER PRIMARY KEY AUTOINCREMENT, movie_id INTEGER, genre_id INTEGER)')
c.execute('CREATE TABLE IF NOT EXISTS companies_replace (id INTEGER PRIMARY KEY AUTOINCREMENT, name TEXT)')
c.execute('CREATE TABLE IF NOT EXISTS movieCompanies_replace (id INTEGER PRIMARY KEY AUTOINCREMENT, movie_id INTEGER, company_id INTEGER)')
c.execute('CREATE TABLE IF NOT EXISTS movieCountries_replace (id INTEGER PRIMARY KEY AUTOINCREMENT, movie_id INTEGER, country_code TEXT)')
c.execute('CREATE TABLE IF NOT EXISTS movieLanguages_replace (id INTEGER PRIMARY KEY AUTOINCREMENT, movie_id INTEGER, language_code TEXT)')
conn.commit()
```

```python
c.execute('INSERT INTO casts_replace (id, movie_id, cast_id, character, credit_id, actor_id, order_number) SELECT * FROM casts')
c.execute('INSERT INTO actors_replace (id, gender, name, profile_path) SELECT id, gender, name, profile_path FROM actors')
c.execute('INSERT INTO collections_replace (id, collection_id, name, poster_path, backdrop_path) SELECT id, collection_id, name, poster_path, backdrop_path FROM collections')
c.execute('INSERT INTO movies_replace (id, adult, belongs_to_collection, budget, homepage, movie_id, imdb_id, original_language, original_title, overview, popularity, poster_path, release_date, revenue, runtime, status, tagline, title, vote_average, vote_count) SELECT id, adult, belongs_to_collection, budget, homepage, movie_id, imdb_id, original_language, original_title, overview, popularity, poster_path, release_date, revenue, runtime, status, tagline, title, vote_average, vote_count FROM movies')
c.execute('INSERT INTO genres_replace (id, name) SELECT id, name FROM genres')
c.execute('INSERT INTO movieGenres_replace (id, movie_id, genre_id) SELECT id, movie_id, genre_id FROM movieGenres')
c.execute('INSERT INTO companies_replace (id, name) SELECT id, name FROM companies')
c.execute('INSERT INTO movieCompanies_replace (id, movie_id, company_id) SELECT id, movie_id, company_id FROM movieCompanies')
c.execute('INSERT INTO movieCountries_replace (id, movie_id, country_code) SELECT id, movie_id, country_code from movieCountries')
c.execute('INSERT INTO movieLanguages_replace (id, movie_id, language_code) SELECT id, movie_id, language_code FROM movieLanguages')
conn.commit()
```

```python
c.execute('DROP TABLE casts')
c.execute('ALTER TABLE casts_replace RENAME TO  casts')
c.execute('DROP TABLE actors')
c.execute('ALTER TABLE actors_replace RENAME TO actors')
c.execute('DROP TABLE collections')
c.execute('ALTER TABLE collections_replace RENAME TO collections')
c.execute('DROP TABLE movies')
c.execute('ALTER TABLE movies_replace RENAME TO movies')
c.execute('DROP TABLE genres')
c.execute('ALTER TABLE genres_replace RENAME TO genres')
c.execute('DROP TABLE movieGenres')
c.execute('ALTER TABLE movieGenres_replace RENAME TO movieGenres')
c.execute('DROP TABLE companies')
c.execute('ALTER TABLE companies_replace RENAME TO companies')
c.execute('DROP TABLE movieCompanies')
c.execute('ALTER TABLE movieCompanies_replace RENAME TO movieCompanies')
c.execute('DROP TABLE movieCountries')
c.execute('ALTER TABLE movieCountries_replace RENAME TO movieCountries')
c.execute('DROP TABLE movieLanguages')
c.execute('ALTER TABLE movieLanguages_replace RENAME TO movieLanguages')
conn.commit()
```
